### These techniques include:

#### 1.   **Data Preprocessing**

*   When preprocessing data for "Text Classification" algorithms, key steps include:

1.   Tokenization (splitting text into individual words).
2.   Lowercase conversion (making all letters lowercase).
3.   Removing punctuation and special characters.
4.   Stop word removal (eliminating common words like "the" and "a").
5.   Stemming or lemmatization (reducing words to their root form).
6.   Expanding contractions (converting shortened words like "can't" to "cannot") to ensure consistent representation of the text data before feeding it into a classification model.

*   When preprocessing data for "Audio Classification" algorithms, key steps include:
1.  Converting raw audio into spectrograms.
2.  Normalizing the audio signal.
3.  Removing background noise.
4.  Extracting relevant features (such as Mel Frequency Cepstral Coefficientsâ€”MFCCs).
5.  Resampling to a consistent sample rate.
6.  Trimming audio segments to a fixed duration.

*The above preparation effectively utilizes the data for machine learning models, enabling them to classify different text and audio types.*

#### 2.   **Feature Engineering Extraction**

In text classification, feature extraction algorithms convert raw text data into numerical features that machine learning models can understand, often using techniques like "Bag of Words" (BoW), TF-IDF (Term Frequency-Inverse Document Frequency), and word embeddings (such as Word2Vec or GloVe) to represent the content of a document as a vector of numerical values. This enables classification based on these features.

When it comes to audio classification, standard feature extraction algorithms include Zero-Crossing Rate (ZCR), Spectral Centroid, Mel-Frequency Cepstral Coefficients (MFCC), Spectral Flux, Spectral Rolloff, Short-Time Energy, and Chroma Features. These algorithms transform raw audio signals into meaningful representations that can be utilized for classification tasks by capturing pitch, timbre, and loudness based on the audio signal's frequency content.

#### 3.   **Model Selection and Evaluation**

In the context of text classification, algorithms refer to the process of choosing the best machine learning model from a set of options by comparing their performance on a held-out validation dataset, using metrics designed explicitly for text classification tasks to determine which model best suits the given data and the problem at hand. It involves trying different algorithms, evaluating their accuracy on unseen data, and selecting the best-performing one.

When selecting and evaluating audio classification algorithms, key considerations include choosing appropriate feature extraction techniques, such as Mel Frequency Cepstral Coefficients (MFCCs) or spectrograms, and comparing the performance of various machine learning models.

**Standard Algorithms:** Naive Bayes, Support Vector Machines (SVM), Logistic Regression, Decision Trees, Random Forests, eXtreme Gradient Boosting (XGBoost), and Neural Networks (CNNs and RNNs).

***CNN Architectures***:LeNet-5, AlexNet, VGG-16, Inception-v1, Inception-v3, ResNet-50, Xception, Inception-v4, Inception-ResNets, ResNeXt-50.

***RNN Architectures***: Fully recurrent neural networks (FRNN,) Hopfield Elman networks and Jordan networks, Long short-term memory (LSTM), Gated Recurrent Unit (GRU), Bidirectional Associative Memory (BAM), Echo State Networks (ESN), Recursive Neural Network (ReNN), Neural Turing Machines (NTMs).

```python
def analyze_dataframe(filepath):
    """Analyzes a CSV file to identify numerical and categorical columns."""
    try:
        df = pd.read_csv(filepath)
        numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
        categorical_cols = df.select_dtypes(exclude=['number']).columns.tolist()

        return {
            'numerical_cols': numerical_cols,
            'categorical_cols': categorical_cols,
            'num_numerical_cols': len(numerical_cols),
            'num_categorical_cols': len(categorical_cols)
        }

    except FileNotFoundError:
        print(f"Error: File not found at '{filepath}'")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

filepath = '/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings/overview-of-recordings.csv'
results = analyze_dataframe(filepath)

if results:
    print("Numerical Columns:")
    for col in results['numerical_cols']:
      print(col)
    print(f"\nNumber of Numerical Columns: {results['num_numerical_cols']}")

    print("\nCategorical Columns:")
    for col in results['categorical_cols']:
      print(col)
    print(f"\nNumber of Categorical Columns: {results['num_categorical_cols']}")
```

#### 3.3.2 Variables Measurements

```python
def analyze_csv_datatypes(filepath):
    """Analyzes a CSV file to determine the data types of each column and categorize them."""
    try:
        df = pd.read_csv(filepath)

        # Categorize columns as numeric or categorical
        numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
        categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

        return {
            "numeric": numeric_cols,
            "categorical": categorical_cols,
            "all_dtypes": df.dtypes
        }

    except FileNotFoundError:
        print(f"Error: File not found at '{filepath}'")
        return None
    except pd.errors.ParserError:
        print(f"Error: Could not parse the CSV file at '{filepath}'. Check the file format.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

filepath = '/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings/overview-of-recordings.csv'
analysis_result = analyze_csv_datatypes(filepath)

if analysis_result:
    print("Data types of each column:")
    print(analysis_result["all_dtypes"])
    print("\nNumeric columns:")
    print(analysis_result["numeric"])
    print("\nCategorical columns:")
    print(analysis_result["categorical"])

def identify_variables(filepath):

    try:
        df = pd.read_csv(filepath)

        # Crude heuristic: all columns before the second to last independent variable

        independent_variables = df.columns[:-2].tolist()

        return {
            "independent_variables": independent_variables
        }
    except FileNotFoundError:
        print(f"Error: File not found at '{filepath}'")
        return None
    except pd.errors.ParserError:
        print(f"Error: Could not parse the CSV file at '{filepath}'. Check the file format.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

filepath = '/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings/overview-of-recordings.csv'
variable_analysis = identify_variables(filepath)

if variable_analysis:
    print("Independent Variables:", variable_analysis["independent_variables"])
```

#### 3.3.3 Categorical Variables Scale

#### 3.3.5 Using the NLTK tokenizer to convert the phrases to integer sequences for Text Features (Tokenization)

```python
def clean_txt(docs):
    lemmatizer = WordNetLemmatizer()
    # split into words
    speech_words = nltk.word_tokenize(docs)
    # convert to lower case
    lower_text = [w.lower() for w in speech_words]
    # prepare regex for char filtering
    re_punc = re.compile('[%s]' % re.escape(string.punctuation))
    # remove punctuation from each word
    stripped = [re_punc.sub('', w) for w in lower_text]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words
    words = [w for w in words if not w in list(STOP_WORDS)]
    # filter out short tokens
    words = [word for word in words if len(word) > 2]
    # Stemm all the words in the sentence
    lem_words = [lemmatizer.lemmatize(word) for word in words]
    combined_text = ' '.join(lem_words)
    return combined_text

# Cleaning the text data
df_text.loc[:, 'cleaned_phrase'] = df_text['phrase'].apply(clean_txt)
df_text
```

### 8.5.1 Split Long Audio Files into Smaller, and Manageable Segments (Audio Segmentation)

```python
def split_audio(input_file, output_dir, segment_duration=10):
    """
    Splits a long audio file into smaller segments.

    Args:
        input_file: Path to the input WAV file.
        output_dir: Directory to save the output segments.
        segment_duration: Duration of each segment in seconds.
    """
    try:
        y, sr = librosa.load(input_file, sr=None)  # Load audio with original sample rate
    except FileNotFoundError:
        print(f"Error: Input file '{input_file}' not found.")
        return
    except Exception as e:
        print(f"Error loading audio file: {e}")
        return

    os.makedirs(output_dir, exist_ok=True)  # Create output directory if it doesn't exist
    filename = os.path.splitext(os.path.basename(input_file))[0]

    total_duration = librosa.get_duration(y=y, sr=sr)
    num_segments = int(total_duration // segment_duration)

    for i in range(num_segments):
        start = i * segment_duration * sr
        end = min((i + 1) * segment_duration * sr, len(y))
        segment = y[int(start):int(end)]
        output_file = os.path.join(output_dir, f"{filename}_segment_{i + 1}.wav")
        sf.write(output_file, segment, sr)

    # Handle any remaining audio that doesn't fit into a full segment
    if total_duration % segment_duration != 0:
        start = num_segments * segment_duration * sr
        segment = y[int(start):]
        output_file = os.path.join(output_dir, f"{filename}_segment_{num_segments + 1}.wav")
        sf.write(output_file, segment, sr)

    print(f"Audio file '{input_file}' split into segments and saved to '{output_dir}'.")

input_wav_file = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings"
subfolders = ['train', 'test', 'validate']
output_directory = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings/segments"

# Loop through subfolders and audio files
for subfolder in subfolders:
    subfolder_path = os.path.join(input_wav_file, subfolder)
    for filename in os.listdir(subfolder_path):
        if filename.endswith(".wav"):
            input_file_path = os.path.join(subfolder_path, filename)
            split_audio(input_file_path, output_directory)
```

### 8.5.2 Adjust the Amplitude of Audio Signals to a Consistent level (Audio Normalization)

```python
def normalize_audio(input_file, output_file, target_dBFS=-20):
    """
    Normalizes the amplitude of an audio signal to a target dBFS level.

    Args:
        input_file: Path to the input WAV file.
        output_file: Path to save the normalized WAV file.
        target_dBFS: Target dBFS level for normalization.
    """
    try:
        y, sr = librosa.load(input_file, sr=None)  # Load audio with original sample rate
        peak_amplitude = np.max(np.abs(y))
        if peak_amplitude == 0:
            print(f"Warning: Input file '{input_file}' has zero peak amplitude. Skipping normalization.")
            return
        gain = target_dBFS - 20 * np.log10(peak_amplitude)
        normalized_audio = y * (10**(gain / 20))
        sf.write(output_file, normalized_audio, sr)
        print(f"Audio file '{input_file}' normalized and saved to '{output_file}'.")
    except FileNotFoundError:
        print(f"Error: Input file '{input_file}' not found.")
    except Exception as e:
        print(f"An error occurred during normalization: {e}")

# Example usage (iterating through subfolders)

data_dir = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings"
# Replace with the actual path to your data directory

subfolders = ['train', 'test', 'validate']  # Add any other relevant subfolders

for subfolder in subfolders:
    input_dir = os.path.join(data_dir, subfolder)  # Path to the input subfolder
    output_dir = os.path.join(data_dir, subfolder + '_normalized')  # Path to the output subfolder

    os.makedirs(output_dir, exist_ok=True)  # Create the output subfolder if it doesn't exist

    for filename in os.listdir(input_dir):
        if filename.endswith(".wav"):
            input_file = os.path.join(input_dir, filename)
            output_file = os.path.join(output_dir, filename)
            normalize_audio(input_file, output_file)
```

### 8.5.3 Convert Audio Waveforms into Spectrograms with methods Short-Time Fourier Transform (STFT) Method (Feature Extraction)

```python
def create_spectrogram(audio_file, output_file, n_fft=2048, hop_length=512):
    """
    Converts an audio file into a spectrogram image using STFT.

    Args:
        audio_file: Path to the input WAV file.
        output_file: Path to save the spectrogram image.
        n_fft: Size of the FFT window.
        hop_length: Number of samples between successive frames.
    """
    try:
        y, sr = librosa.load(audio_file, sr=None)  # Load audio with original sample rate
        stft_result = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
        spectrogram = np.abs(stft_result)  # Magnitude spectrogram

        # Convert to decibels (dB) for better visualization
        db_spectrogram = librosa.amplitude_to_db(spectrogram, ref=np.max)

        # Create and save the spectrogram image
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(db_spectrogram, sr=sr, hop_length=hop_length, x_axis='time', y_axis='log')
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'Spectrogram of {os.path.basename(audio_file)}')
        plt.savefig(output_file)
        plt.close()  # Close the plot to free up memory
        print(f"Spectrogram saved to {output_file}")
    except FileNotFoundError:
        print(f"Error: Input file '{audio_file}' not found.")
    except Exception as e:
        print(f"An error occurred during spectrogram creation: {e}")


# Example usage (iterating through subfolders):

data_dir = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings"

subfolders = ['train_normalized', 'test_normalized', 'validate_normalized']
spectrogram_output_dir = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/spectrograms"

os.makedirs(spectrogram_output_dir, exist_ok = True) # Create output directory if it doesn't exist


for subfolder in subfolders:
  input_dir = os.path.join(data_dir, subfolder)
  output_subdir = os.path.join(spectrogram_output_dir, subfolder) # Create subdirectory for spectrograms
  os.makedirs(output_subdir, exist_ok=True)

  for filename in os.listdir(input_dir):
      if filename.endswith(".wav"):
          input_file = os.path.join(input_dir, filename)
          output_file = os.path.join(output_subdir, filename[:-4] + ".jpg") # Save as JPG
          create_spectrogram(input_file, output_file)
```

### 8.5.4 Capture Important Spectral Characteristics with Mel Frequency Cepstral Coefficients (MFCC)

```python
def extract_mfccs(audio_file, num_mfccs=13):
    """
    Extracts Mel-frequency cepstral coefficients (MFCCs) from an audio file.

    Args:
        audio_file: Path to the input WAV file.
        num_mfccs: The number of MFCCs to extract.

    Returns:
        A tuple containing MFCCs (NumPy array) and sample rate (int).
    """
    try:
        y, sr = librosa.load(audio_file, sr=None)
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=num_mfccs)
        return mfccs, sr # Return sr along with mfccs
    except FileNotFoundError:
        print(f"Error: Input file '{audio_file}' not found.")
        return None, None # Return None for both if error
    except Exception as e:
        print(f"An error occurred during MFCC extraction: {e}")
        return None, None # Return None for both if error

def display_mfccs(mfccs, sr, audio_file): # Add sr as argument
    """
    Displays the extracted MFCCs as an image.
    """
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfccs, sr=sr, x_axis='time') # Use the passed sr
    plt.colorbar()
    plt.title(f'MFCCs of {os.path.basename(audio_file)}')
    plt.tight_layout()
    plt.show()

# Example usage
data_dir = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings"
subfolders = ['train_normalized'] # Process one subfolder as an example
for subfolder in subfolders:
    input_dir = os.path.join(data_dir, subfolder)
    for filename in os.listdir(input_dir):
        if filename.endswith(".wav"):
            audio_file = os.path.join(input_dir, filename)
            mfccs, sr = extract_mfccs(audio_file) # Get sr from extract_mfccs
            if mfccs is not None:
                display_mfccs(mfccs, sr, audio_file) # Pass sr to display_mfccs
```

### 8.5.5 Apply Transformations (Audio Data Augmentation)

```python
def apply_transformations(audio_file, output_file, transformations):
    """
    Applies a series of audio transformations to an audio file.

    Args:
        audio_file: Path to the input WAV file.
        output_file: Path to save the transformed WAV file.
        transformations: A list of dictionaries, where each dictionary defines
                         a transformation to apply.  Each dictionary must have
                         a 'type' key and any other parameters needed for the
                         specific transformation.
                         Supported transformation types: 'pitch_shift', 'time_stretch', 'noise'
    """
    try:
        y, sr = librosa.load(audio_file, sr=None)
        for transform in transformations:
            if transform['type'] == 'pitch_shift':
                y = librosa.effects.pitch_shift(y, sr, n_steps=transform.get('n_steps', 2))
            elif transform['type'] == 'time_stretch':
                y = librosa.effects.time_stretch(y, rate=transform.get('rate', 1.1))
            elif transform['type'] == 'noise':
                noise_amplitude = transform.get('noise_amplitude', 0.005)
                noise = np.random.randn(len(y)) * noise_amplitude
                y = y + noise
            else:
                print(f"Warning: Unsupported transformation type: {transform['type']}")
        sf.write(output_file, y, sr)
        print(f"Audio transformations applied to '{audio_file}' saved to '{output_file}'.")
    except FileNotFoundError:
        print(f"Error: Input file '{audio_file}' not found.")
    except Exception as e:
        print(f"An error occurred during audio transformations: {e}")


# Example transformations
transformations_list = [
    {'type': 'pitch_shift', 'n_steps': 2},
    {'type': 'time_stretch', 'rate': 1.1},
    {'type': 'noise', 'noise_amplitude': 0.005}
]


# Example usage (applying to all wav files in a subdirectory)
data_dir = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/recordings"
subfolders = ['train_normalized']
transformed_output_dir = "/content/drive/MyDrive/ai_env/projects/Text and Audio Classification Enabled Diagnosis for Treatment Applications by NLP and DL/Medical Speech, Transcription, and Intent/transformed_audio"


import os
os.makedirs(transformed_output_dir, exist_ok=True) # Create output directory if it does not exist


for subfolder in subfolders:
    input_dir = os.path.join(data_dir, subfolder)
    output_subdir = os.path.join(transformed_output_dir, subfolder) # Create subdirectory for transformed audio
    os.makedirs(output_subdir, exist_ok=True)

    for filename in os.listdir(input_dir):
      if filename.endswith(".wav"):
          input_file = os.path.join(input_dir, filename)
          output_file = os.path.join(output_subdir, filename)
          apply_transformations(input_file, output_file, transformations_list)
```
